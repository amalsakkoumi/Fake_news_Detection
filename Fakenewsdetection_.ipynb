```python
def wordopt(text):
    # convert to a lower case
    text = text.lower()
    # remove  Urls
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # remove html tags
    text = re.sub('<.*?>+', '', text)
    # remove punctuation
    text = re.sub(r'[^\w\s]','',text)

    # Remove digits
    text = re.sub(r'\d', '', text)
    # Remove newline characters
    text = re.sub('\n', '', text)
    return text
```
